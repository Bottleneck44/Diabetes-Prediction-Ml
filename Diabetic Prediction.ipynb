```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# Data Analysis 

### PIMA DATASET(KAGGLE)

```python
df = pd.read_csv(r"C:\Users\Ananda\Desktop\PYTHON\machine learning projects\Diabetes Prediction\diabetes.csv")
```

```python
df.describe()
```

```python
df['Outcome'].value_counts()
```

## Seperating labels and values

```python
X = df.drop(columns='Outcome',axis=1)
Y = df['Outcome']
```

```python
X
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
stdata = scaler.transform(X)
```

```python
X = stdata
```

### Train Test Split

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,stratify=Y,test_size=0.2,random_state=2)
```

## Training The Model

```python
classifier = svm.SVC(kernel='linear')
```

## Training the support vector machine

```python
classifier.fit(X_train,Y_train)
```

### Model Evaluation

```python
# evaluating the model
xtra = classifier.predict(X_train)
xtrda = accuracy_score(xtra,Y_train)
print(xtrda)
```

```python
xted = classifier.predict(X_test)
xteda = accuracy_score(xted,Y_test)
print(xteda)
```

## Making a Predictive System

```python
input_data = (1,85,66,29,0,26.6,0.351,31)
inpar = np.asarray(input_data)
inp = inpar.reshape(1,-1)
std_data = scaler.transform(inp)
pred = classifier.predict(std_data)
if pred == 1:
    print('Diabetic')
else:
    print('Non-Diabetic')
```
